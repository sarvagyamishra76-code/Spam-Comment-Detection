# NLP Preprocessing for Spam Detection

```
import pandas as pd
import nltk, string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
df = pd.read_csv('cleaned_spam.csv')
ps = PorterStemmer()
def transform_text(text):
    text=str(text).lower()
    words=text.split()
    return ' '.join(ps.stem(w) for w in words if w not in stopwords.words('english'))
df['transformed_text']=df['text'].apply(transform_text)
df.head()
```
